Parch | Number of Parents / Children Aboard

Fare | Passenger Fare
Cabin | Cabin
Embarked | Port of Embarkation (C =Cherbourg; Q=Queenstown; S=Southampton)

### **5. DT Predictive Modelling & Visualization**

```python
# Import decision tree from scikit-learn module
from sklearn import tree
```

```python
train_data = pd.DataFrame()  # Creates a new empty DataFrame
train_data['Age'] = titanic_df['Age']  # Adds only the 'Age' column from df
```

```python
# Converting Sex data to numerical data (1 = male, 0 = female)
train_data['Sex'] = pd.Series([1 if s == 'male' else 0 for s in titanic_df.Sex], name='Sex')
train_labels = titanic_df['Survived']
```

```python
# Build a DT classifier
classifier = tree.DecisionTreeClassifier(max_depth=3)
classifier.fit(train_data, train_labels)
```

```python
# Performance of classifier
print(f'Score: {classifier.score(train_data, train_labels)}')
```

```python
# To predict survivability in Titanic. Given age 30 and 0 = female
print('30 years old female: ', classifier.predict([[30, 0]]))

# To predict survivability in Titanic. Given age 30 and 1 = male
print('30 years old male : ', classifier.predict([[30, 1]]))
```
